# Part I: Research Question

## Context 
The contribution of the study to the field of data analytics and the MSDA program is to create a predictive model which can estimate the price of cars in the Indian used car market, so that a company does not need to rely off estimation and can help used card dealers and online car dealers predict potential profits for cars.  There is a huge demand for used cars in the Indian Market today. As sales of new cars have slowed down in the recent past, the pre-owned car market has continued to grow over the past years and is larger than the new car market now. 

In 2018-19, while new car sales were recorded at 3.6 million units, around 4 million second-hand cars were bought and sold. There is a slowdown in new car sales and that could mean that the demand is shifting towards the pre-owned market. In fact, some car sellers replace their old cars with pre-owned cars instead of buying new ones. Unlike new cars, where price and supply are fairly deterministic and managed by Original Equipment Manufacturer, used cars are very different beasts with huge uncertainty in both pricing and supply. 

Keeping this in mind, the pricing scheme of these used cars becomes important to grow in the market. A pricing model that can effectively predict the price of used cars would greatly aid the business in devising profitable strategies using differential pricing (Kausthub, 2021). Both Kaushal (2021) and Kaustub found that using multiple linear regression methods, an effective mode can be created that can accurately predict prices, given both continuous and categorical variables. 

### Project Topic
Multiple linear regression analysis of Indian used card prices, and principal component analysis of features affecting price, segmented by brand

### Research Question
Can a regression model be created from the market research data set?

### Hypothesis
* Null hypothesis- A multiple linear regression model cannot be created from the market research dataset. 
* Alternate Hypothesis- - A multiple linear regression model can be created from the market research dataset

## Project Outcomes
The project will seek to create a multiple linear regression model for the data, using the most important features. The model will prune features down using Principal Component Analysis to select the most important features for the car pricing model, based off thresholds. Multiple linear regression will be used to estimate the pricing for used cars in India. Shapley Additive Explanations (SHAP values) will be used to determine feature importance of the top 5 used cars to determine feature importance. Support for the alternative hypothesis is found in Kausthub (2021) and Kaushal (2021), that a multiple linear regression model can be helpful in estimating sales prices by determining the strength of the features with respect to sale price.

# Part III: Data Extraction and Preparation

### Data Extraction and Preparation Methods

Data extraction had two phases: initial extraction and preparation of the data for model. Intial extraction was conducted using the Pandas library, while data preparation was done using several different libraries, including missingno, pandas, and numpy. The goal of the extraction and preperation was to understand the dataset, perform feature engineering, and clean up dirty data prior to modeling. 

Extraction was the first phase, and focused on loading the data into a dataframe. Data was extracted from a csv file downloaded from the Kaggle website that contained the data scraped from Cardekho.com. The data was extracted using the read_csv method from the csv, then checked for the size and descriptive statistics of the data. This showed that the data had over 20,000 rows with 16 features, the latter of which would be transformed in the preparation.

Peparation focused on cleaning dirty data and feature engineering for modeling. Feature engineering and data type reformatting focused on:
* Dropping the columns web-scraper-order, web-scraper-start-url, Source.Name, new-price which were not needed for the model.
* Changing Selling Price into Selling Price Ladkhs, then changing the data type from string to float.
* Extracting Car Name from Car Full Name, and dropping Car Full Name from the dataframe.
* Creating a new feature called car age, which subtracted the current year from the manufacturing year of the car
* Extracting number kilometers driven from km_driven column, then changing the data type to float from string. 
* Extracting the number of seats from the Seats column, creating an new column, then changing the data type to float from string. The original seats column was then dropped. 
* Extracting the milage number from the milage column, then changing the data type to float from string 
* Getting the car brand. Car name, which was created eariler in the process, was broken up into two new columns: Car Brand and Car Model. Car brand names were then standardized to improve consistency of the data.
* Encoding owner_type, fuel_type, and transmission_type as categorical variables. This changed the data type from a string data type to a categorical one.

Once the data had been transformed, it was then checked for zero values. From there, the columns km_driven, max_power, car_name, and current_year were dropped since their values had been already transformed as mentioned previously. Columns were then reordered for readability when performing exploratory data analysis and model creation.


## Test Assumptions

To test, we will check the assumptions of linear regression. There are four checks to ensure that there is the model is valid: No Multicollinearity, Mean of residuals should be 0, No Heteroscedasticity, Linearity of variables, and Normality of error terms. These are essential to proving wether or not the machine learning model can be used in production (Zheng 240).


### Checking Assumption 1: No Multicollinearity 

We will use VIF, to check if there is multicollinearity in the data.

Features having a VIF score >5 will be dropped/treated till all the features have a VIF score <5

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


def checking_vif(train):
    vif = pd.DataFrame()
    vif["feature"] = train.columns

    # calculating VIF for each feature
    vif["VIF"] = [
        variance_inflation_factor(train.values, i) for i in range(len(train.columns))
    ]
    return vif
```

```python
# Check VIF
print(checking_vif(X_train))
```

### Checking Assumption 3: No Heteroscedasticity


* Homoscedacity - If the residuals are symmetrically distributed across the regression line , then the data is said to homoscedastic.

* Heteroscedasticity- - If the residuals are not symmetrically distributed across the regression line, then the data is said to be heteroscedastic. In this case the residuals can form a funnel shape or any other non symmetrical shape.

We'll use `Goldfeldquandt Test` to test the following hypothesis

Null hypothesis : Residuals are homoscedastic
Alternate hypothesis : Residuals have hetroscedasticity

alpha = 0.05 

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip

name = ["F statistic", "p-value"]
test = sms.het_goldfeldquandt(residuals, X_train)
lzip(name, test)
```


    
Since p-value > 0.05 we cannot reject the Null Hypothesis that the residuals are homoscedastic.

## Summary of Data Analysis 

## Analysis Summary

The data analysis is broken down into four parts. The first is the exploratory data analysis to check distributions and relationships. The second part is to balance out imbalanced data sets. The third part is the creation of baseline model using ordinary least squares regression. The fourth part is the validation of the model and feature selection. 

Analysis of the data set began with an exploratory data analysis to check univariate distributions and then bivariate distributions. This is to determine the range of the data to see if features are imbalanced, and to look at relationships between variables. The other purpose of this is to determine if data is missing and relationships needed to create an effective model. Features were examined to see if distrubutions were skewed, in order to apply transformations to rebalance it. Rebalancing is necessary to create a balanced data set for a model would generate higher accuracy models, higher balanced accuracy and balanced detection rate (Zheng 102). The purpose of the EDA is to check for possible data issues and to optimize the model. 

The next stage was reblancing several columns in the data set. Since kilometeres driven and selling price in ladks were right skewed, we used log transformation to rebalance the data set. Log transformation is used to normalize the data. When our original continuous data do not follow the bell curve, we can log transform this data to make it as “normal” as possible so that the statistical analysis results from this data become more valid (Lakshmanan 222). The point of log transformation is to reduce or remove the skewness of our original data.

Once we have normalized the data, then we create the model. Ordinary least squares regression was used to for estimating the coefficients of the linear regression equation. We picked the simplest regression model, which is different than the original proposal. This is for several reasons. The first is that a simple model is easier to improve upon. More complex means of regression such as lasso, ridge or decision tree regression require that a baseline OLS model is able to be created. The second is that a simpler model is easier to deploy into a production environment first. This allows both data scientists and machine learning engineers to effectively debug the model and test it with real world data. Models that work well with sampled data, don't always work well with real data. To achieve effective turnaround, a simpler model is best. 

Validation is the next step. Validation is important to check if the model can even be used. In this model, four validations were used: checking for multicollinearity, mean of resduals is zero, heteroscedacity, linearity of variables, and normality of error terms. To check for multicolinearity, we used the variance inflation score (VIF) to see how correlated features are. Mean of residuals are checked to be close to zero, meaning that the summation of all predicted value minus the observed value are zero or approach zero. This means that the model has good fit for the regression line. Hetrosedacity was checked using the Goldfeldquandt test, to check the symmetry of the residuals across the regression line. Next linearity of residual vs fitted values was checked. And finally the normality of error terms was checked to see if residuals were evenly disbuted. If the outputs from the model failed this test, it is not possible to create a useful model. 



## Analysis Advantages 

The advantages of the analysis were primarily two: normalization of skewed values and validation checks for the model. The intial exploratory data analysis showed that prices for the target variable, selling price were skewed right. This meant that the distribution of data was closer to a lower price than a higher price. The normalization using log transformation was useful in not only improving the accuracy of the model, but allowing a useful model that could be used in production. Validation checks also played a major advantage in this analysis. These confirmed the usefulness of the normalization of the right skewed selling price values, allowing a rudimentary multiple linear regression model to be built. 


## Analysis Disadvantages

The disadvantages were two: lack of dates when the car was sold and extreme outliers that were present after normalization. 

The lack of dates when each individual car was sold makes it difficult to build a reliable model. It is unknown when the car was sold, which is important for determining trends in the data. This combined with car model year and type is very useful to see what sort of cars and years are selling well, and also to segement for cars. The lack of dates when the car was sold also could be leading to data leakage. 

Data leakage occurs when information from outside the training dataset is used to create the model. In this model's case, it would be mixing different sales dates for the cars. The training and testing data set would sampling from dates not in the same time period. This could result in the use of information in the model training process which would not be normally availible during the model training time. For example, training data of selling prices in ladkhs from 2019 being mixed with the selling prices in 2020. This would result in a model that is not only over accurate, but also one that is using future data to predict. This results in an extremely biased and unreliable model. When data leakage occurs, leads to overly optimistic outcomes during the model building phase, followed by poor results after the prediction model is implemented and tested on new data in production. 

Another disadvantage was the extreme outliers after normalization. Normalization is usually used with imbalanced or skewed datasets to create a normalized distribution of data. In this case, the data that was being normalized was the selling price in ladkhs. However, even after normalization using log transform, variables remained non-linear during model validation. This suggests that there might be more issues within the data, even after using log transform to normalize it. This will require further study, since it is inadvisable to remove outliers without examining them first. 

Extreme outliers in the target feature (selling price ladkhs) and the lack dates, suggest that further analysis of the data set and extra data will be needed to make this an effective model.

The data in the current form does not meet the quality standards needed for a machine learning model to deliver business values. There are several glaring problems that prevent it's usefulness:
